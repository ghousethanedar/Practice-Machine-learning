```python
# Extracting the Titles from Name....
# Step01: Split the Names into Surname and First Name
newnames = tita["Name"].apply(lambda x:x.split(", ")[-1])

# Extracting the Titles from Name....
# Step02: Split the Names into Title and First Name
titles = newnames.apply(lambda x: x.split(". ")[0])

titles.unique()

# titles that are to be kept - Mr, Mrs, Miss, Master & Others
titles_ignore =['Don', 'Rev', 'Dr', 'Mme', 'Ms',
       'Major', 'Lady', 'Sir', 'Mlle', 'Col', 'Capt', 'the Countess',
       'Jonkheer'] 

# replacing titles_ignore with Others
titles.replace(to_replace=titles_ignore, value="Others", inplace=True)

titles.unique()
```

```python
tita['Survived']=tita['Survived'].astype('str')
tita['Age']=tita['Age'].astype('int64')
tita['Pclass']=tita['Pclass'].astype('str')
```

```python
# Size of the Family

def size_family(x):
    if (x==1):
        return(0)
    elif(x>1) and (x<=4):
        return(1)
    elif(x>4):
        return(2)
    
# Applying the Function to the Family Variable:: Train Set
tita["Family Size"]=tita["Family"].apply(size_family)
```

```python
# Category: Dependent Column:
# 0. Kids Till 10
# 1. Teenager Till 19
# 2. Adults > 19
# 3. Old Aged > 50

def dependent(x):
    if (x<=10):
        return(0)
    elif (x<=19):
        return 1
    elif (x<=50):
        return 2
    else:
        return 3
    
# Applying Dependent Function on Age:: Train Set
tita["Dependents"]=tita["Age"].apply(dependent)
```
